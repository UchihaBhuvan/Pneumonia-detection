```python
import tensorflow
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
DataDir = r"ChestXrays/train"
```

```python
CATEGORIES = ["normal", "opacity"]
```

```python
for i in CATEGORIES:
    path = os.path.join(DataDir, i)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break
    break
```

```python
img_size = 100
```

```python
new_array = cv2.resize(img_array, (img_size, img_size))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
training_data = []
```

```python
def create_training_data():
    for i in CATEGORIES:
        path = os.path.join(DataDir,i)
        class_num = CATEGORIES.index(i)
        
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
import random
random.shuffle(training_data)
for sample in training_data[:10]:
    print(sample)
```

```python
X = []
y = []
```

```python
for features, label in training_data:
    X.append(features)
    y.append(label)
```

```python
print(X[0].reshape(-1,img_size,img_size,1))
```

```python
y = np.array(y)
```

```python
X = np.array(X).reshape(-1,img_size,img_size,1)
```

```python
validation_data=[]
DataDir_val = r"ChestXrays/val"
```

```python
def create_validating_data():
    for i in CATEGORIES:
        path = os.path.join(DataDir_val,i)
        class_num = CATEGORIES.index(i)
        
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array,(img_size,img_size))
                validation_data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_validating_data()
```

```python
print(len(validation_data))
```

```python
import random
random.shuffle(validation_data)
for sample in validation_data[:10]:
    print(sample)
```

```python
X_val = []
y_val = []
```

```python
for features, label in validation_data:
    X_val.append(features)
    y_val.append(label)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation,Flatten
import pickle
from keras.layers import Conv2D, MaxPooling2D
```

```python
X = X.astype('float32')/255.0
X_val = np.array(X_val).reshape(-1,img_size,img_size,1)
X_val = X_val/255.0
y_val = np.array(y_val)
```

```python
X_val.shape
```

```python
model = Sequential()

model.add(Conv2D(64,(3,3),input_shape=X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Dropout(0.2))

model.add(Conv2D(128,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Dropout(0.2))

model.add(Conv2D(256,(3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(64))

model.add(Dropout(.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,batch_size=4,epochs=10,validation_data=(X_val,y_val))
```

```python
model.save(r"model_10.h5")
```

```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.models import load_model
```

```python
CATEGORIES = ["normal","opacity"]
```

```python
def prepare(image):
    img_size=100
    img=tf.keras.preprocessing.image.load_img(image,color_mode="grayscale",target_size=(img_size,img_size))
    new_array = tf.keras.preprocessing.image.img_to_array(img)
    return new_array.reshape(-1,img_size,img_size,1)
```

```python
model = tf.keras.models.load_model(r"model_10.h5")
```

```python
image = 'ChestXrays/test/IM-0087-0001.jpeg'
```

```python
prediction = model.predict([prepare(image)/255.0])
prediction
```

```python
# plt.imshow(prepare(image),cmap="gray")
# plt.show()
print(prediction)
print((prediction[0][0]))
print(CATEGORIES[int(round(prediction[0][0]))])
```
